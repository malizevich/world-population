# World Population Statistics - 2023

**Reference:**

[Bhavik Jikadara. (2023). World Population Statistics 2023. Kaggle](https://www.kaggle.com/datasets/bhavikjikadara/world-population-statistics-2023)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Begin by looking at the source using dataframe

## Exploratory Data Analysis (EDA):

```python
# Calculate total population in 2023
total_population_2023 = df['2023 population'].sum()
print(total_population_2023)
```

```python
# Plot distribution of population based on continents
from matplotlib.ticker import FuncFormatter


plt.figure(figsize=(10, 6))
sns.barplot(x='continent', y='2023 population', data=df, estimator=sum, errorbar=None)
plt.title('Distribution of Population Based on Continents (2023)')
plt.ylabel('Total Population')

# Format y-axis ticks as billions with a thousand separator
def billions_formatter(x, pos):
    return f'{x / 1e9:.1f}B'

plt.gca().yaxis.set_major_formatter(FuncFormatter(billions_formatter))
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```

## Feature Engineering


Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical variable 'continent'
le = LabelEncoder()
df['continent_encoded'] = le.fit_transform(df['continent'])
```
